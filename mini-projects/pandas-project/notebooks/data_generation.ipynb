```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

orders = []

end_date = datetime.today()
start_date = end_date - timedelta(days=365)


for i in range(1000):

    order = {
        "order_id": f"O{i+1}",
        "customer_id": f"C{random.randint(1, 300)}",
        "product_id": f"P{random.randint(1, 100)}",
        "quantity": random.randint(1, 5),
        "price": random.randint(100, 5000),
        "order_date": (start_date + timedelta(days=random.randint(0, 365))).strftime("%Y-%m-%d")
    }

    orders.append(order)

df = pd.DataFrame(orders)

# df.head()
# df.tail()
# df.info() 
# df.describe()
# df.columns
# df.index
# df['order_id']
# df[['order_id','product_id','order_date']]

# Make data messy
df.loc[5, 'price'] = np.nan # Incorrect price
df.loc[10, 'quantity'] = np.nan # Incorrect quantity

df = pd.concat([
    df,
    df.iloc[[0]]
]) # Duplicate data
df.loc[7, 'quantity'] = -5 # Incorrect quantity

df.to_csv(
    '../data/raw/orders.csv',
    index=False
)
```
