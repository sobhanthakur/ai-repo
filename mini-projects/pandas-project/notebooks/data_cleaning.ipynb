```python
import pandas as pd

df = pd.read_csv('../data/raw/orders.csv')

df['price'] = df['price'].fillna(
    df['price'].median()
)

df = df[df['quantity'] > 0] # Negative quantities are removed

df = df.drop_duplicates()

# Create a new column called sales
df['sales'] = (
    df['quantity'] * df['price']
)

df['order_date'] = pd.to_datetime(
    df['order_date'],
    errors='coerce'
)

df.to_csv(
    '../data/cleaned/orders_cleaned.csv',
    index=False
)
```
