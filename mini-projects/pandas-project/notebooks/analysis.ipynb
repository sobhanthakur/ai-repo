```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/cleaned/orders_cleaned.csv')
```

```python
df.head()
```

```python
df['order_date'] = pd.to_datetime(
    df['order_date'],
    errors='coerce'
)
df['month'] = df['order_date'].dt.month

monthly_sales = df.groupby('month')[
    'sales'
].sum()
monthly_sales
```

```python
# Total Revenue
df['sales'].sum()
```

```python
# df['sales'].sum()/df['quantity'].sum()
# Average Order Value (AOV)
df.groupby('order_id')[
    'sales'
].sum().mean()
```

```python
# Highest Sales Day
daily_sales = df.groupby('order_date')[
    'sales'
].sum()

highest_day = daily_sales.idxmax()
print(highest_day)
```

```python
# Weekend vs Weekday Sales
df['day_type'] = np.where(
    df['order_date'].dt.dayofweek >= 5,
    'Weekend',
    'Weekday'
)

sales_by_daytype = df.groupby(
    'day_type'
)['sales'].sum()
print(sales_by_daytype)
```

```python
# Top selling products
top_products = (
    df.groupby('product_id')['sales']
    .sum()
    .sort_values(ascending=False)
)
print(top_products)
```

```python
# Most ordered products
top_products = (
    df.groupby('product_id')['quantity']
    .sum()
    .sort_values(ascending=False)
)
print(top_products)
```

```python
# Low performing product
low_product = (df.groupby('product_id')['sales']
.sum()
.sort_values(ascending=True)
.head(10)
)
print(low_product)
```

```python
# Top customers
top_customers = (
    df.groupby('customer_id')['sales']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
print(top_customers)
```
