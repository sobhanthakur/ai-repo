```python
from transformers import pipeline

# Load pretrained BERT QA model
qa_pipeline = pipeline(
    "question-answering",
    model="bert-large-uncased-whole-word-masking-finetuned-squad"
)

context = """
BERT is a transformer-based language model developed by Google in 2018.
It is widely used for NLP tasks such as text classification,
question answering, and sentiment analysis.
"""

question = "Who developed BERT?"

result = qa_pipeline(
    question=question,
    context=context
)

print("Answer:", result["answer"])
print("Confidence:", result["score"])
```
