```python
from ollama import Client

client = Client()

messages = [
  {
    'role': 'user',
    'content': 'Why is the sky blue?',
  },
]

for part in client.chat('gpt-oss:120b-cloud', messages=messages, stream=True):
  print(part.message.content, end='', flush=True)
```

```python
from ollama import generate

response = generate('gpt-oss:120b-cloud', 'Why is the sky blue?')
print(response['response'])
```

```python
client = Client()

messages = [
  {
    'role': 'user',
    'content': 'Why is the sky blue?',
  },
]

for part in client.chat('gpt-oss:120b-cloud', messages=messages, stream=True, options={"temperature":1}):
  print(part.message.content, end='', flush=True)
```
