```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
import pandas as pd
import re
```

```python
data = pd.read_csv("tripadvisor_hotel_reviews.csv")
```

```python
data.head()
```

```python
data['review_lowercase'] = data['Review'].str.lower()
```

```python
data.head()
```

```python
en_stopwords = stopwords.words('english')
print(en_stopwords)
print(len(en_stopwords))
```

```python
print(len(en_stopwords)) # Gives same result because the list does not contain the word not
```

```python
data['no_stopwords'] = data['review_lowercase'].apply(lambda x: ' '.join([word for word in x.split() if word not in en_stopwords]))
```

```python
data.head()
```

```python
data['review_no_stopwords_no_punct'] = data.apply(lambda x: re.sub(r"[*]", "star", x['no_stopwords']), axis=1)
```

```python
data['review_no_stopwords_no_punct'] = data. \
                                    apply(lambda x: \
                                    re.sub(r"([^\w\s])", \
                                    "", \
                                    x['review_no_stopwords_no_punct'] \
                                    ), axis=1)
```

```python
data.head()
```

```python
# data['tokenized'] = data.apply(lambda x: word_tokenize(x['review_no_stopwords_no_punct']), axis=1)
data['tokenized'] = data['review_no_stopwords_no_punct'].apply(lambda x: word_tokenize(x))
```

```python
data.head()
```

```python
tokens_clean = sum(data['tokenized'],[])
```

```python
print(tokens_clean)
```

```python
unigrams = (pd.Series(nltk.ngrams(tokens_clean,1)).value_counts())
print(unigrams)
```

```python
bigrams = (pd.Series(nltk.ngrams(tokens_clean,2)).value_counts())
print(bigrams)
```
