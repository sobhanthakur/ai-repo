*Some model's sentiment score lies between -1 to 1, where -1 is most negative, 0 is neutral and 1 being most positive (Especially for vader or textblob). Some ML/DL models outputs probabilities like positive being 0.92 and negative being 0.08. 
Some use labels like positive, negative. Some use stars like 1 star (Most Negative) and 5 star being (Most positive)*

`Some transformer use score, which defines confidence probability.
[
  {
    'label': 'POSITIVE',
    'score': 0.998
  }
]`

```python
from textblob import TextBlob
```

```python
sentence_1 = "i had a great time at the movie it was really funny"
sentence_2 = "i had a great time at the movie but the parking was terrible"
sentence_3 = "i had a great time at the movie but the parking wasn't great"
sentence_4 = "i went to see a movie"
```

```python
sentiment_score = TextBlob(sentence_1)
print(sentiment_score.sentiment.polarity)
```

```python
sentiment_score = TextBlob(sentence_2)
print(sentiment_score.sentiment.polarity)
```

```python
print(TextBlob(sentence_3).sentiment.polarity)
```

```python
print(TextBlob(sentence_4).sentiment.polarity)
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
vs = SentimentIntensityAnalyzer()
```

```python
print(vs.polarity_scores(sentence_1))
```

```python
print(vs.polarity_scores(sentence_2))
```

```python
print(vs.polarity_scores(sentence_3))
```

```python
print(vs.polarity_scores(sentence_4))
```
