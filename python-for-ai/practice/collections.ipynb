```python
# Find the Largest Number (Beginner)
numbers = [12, 45, 2, 89, 67]
max_num = -1
for i in numbers:
    if i > max_num:
        max_num = i
print(max_num, max(numbers))
```

```python
# Remove Duplicates While Preserving Order (Easy)
numbers = [2,2,3,1,1,4,5,5]
dict = {}
dedup=[]
for i in numbers:
    if i in dict:
        dict[i] += 1
    else:
        dict[i] = 1
for key, value in dict.items():
    dedup.append(key)
print(dedup)
```

```python
# Group Words by Length (Medium)
list = ["cat","apple","dog","banana"]
dict = {}
for i in list:
    if len(i) in dict:
        dict[len(i)].append(i)
    else:
        dict[len(i)] = [i]

print(dict)
```

```python
# Find the Second Largest Number (Medium)
list = [10,5,20,8,18]
list.sort()
max_2 = max(list[:-1])
max_2
```

```python
# Rotate a List by k Positions
numbers = [1,2,3,4,5]
k = 2
last = numbers[-k:]
first = numbers[:-k]
# [*last,*first]
last + first # Merges array
```

```python
# Find Common Elements Between Two Lists (Hard)
a = [1,2,3,4,5]
b = [3,5,7,8]
s1 = set()
for i in a:
    if i in b:
        s1.add(i)

for i in b:
    if i in a:
        s1.add(i)
print(s1)
```

```python
# Build a Student Marks System (Advanced)
students = [
    {"name":"Alice","marks":90},
    {"name":"Bob","marks":75},
    {"name":"Charlie","marks":85}
]

totalMarks = 0
count = 0

for i in students:
    totalMarks += i['marks']
    count += 1

totalMarks/count
```
