```python
# With Condition
numbers = [x for x in range(1,11)]
[x for x in numbers if x % 2 == 0]
```

```python
# Starts with letter 'a' and ends with 'y'
words = ["a","","alley","angry","applying","array","ash"]
[word for word in words if word and word[0] == "a" and word[-1] == "y"]
```

```python
# Starts with letter 'a' and ends with 'y'
words = ["a","","alley","angry","applying","array","ash"]
[word for word in words if len(word) > 1 and word.startswith("a") and word.endswith("y")]
```

```python
# Flatten an array
# list comprehension with 2 for loops
matrix=[[1,2,3],[4,5],[6,7]]
[inner for outer in matrix for inner in outer]
```

```python
# Categorize a number as even or odd
[f"Even {x}" if x%2 == 0 else f"Odd {x}" for x in range(1,11)]
```

```python
marks = [95, 82, 74, 61, 48]

grades = [
    f"{m} = A" if m >= 90
    else f"{m} = B" if m >= 80
    else f"{m} = C" if m >= 70
    else f"{m} = D" if m >= 60
    else f"{m} = F"
    for m in marks
]

print(grades)
```
