```python
from langchain_community.document_loaders import PyPDFLoader
import copy
```

```python
loader_pdf = PyPDFLoader("Introduction_to_Data_and_Data_Science.pdf")
```

```python
pages_pdf = loader_pdf.load()
```

```python
pages_pdf # Prints the metadata
```

```python
pages_pdf_cut = copy.deepcopy(pages_pdf)
```

```python
' '.join(pages_pdf_cut[0].page_content.split()) # Shows the first page
```

```python
# This cleans the text.
# Remove extra spaces
# Remove line breaks (\n)
# Remove tabs (\t)
# Normalize all whitespace to single spaces
# Joins the words back together using a single space.

for i in pages_pdf_cut:
    i.page_content = ' '.join(i.page_content.split())
```
