```python
from langchain_core.runnables import chain
```

```python
@chain
def runnable_sum(x):
    return sum(x)

@chain
def runnable_square(x):
    return x**2
```

```python
type(runnable_sum), type(runnable_square)
```

```python
chain = runnable_sum | runnable_square
```

```python
chain.invoke([1, 2, 5])
```
